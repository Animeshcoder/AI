### **Introduction to BFS:**

Breadth First Search (BFS) is an algorithm used for traversing or searching tree or graph data structures. The process starts at the root (selecting some arbitrary node as the root in the case of a graph) and explores all the neighboring nodes before moving on to the next level. This means that, given a starting node, BFS will first explore all the nodes that are one edge away from the starting node, then all the nodes that are two edges away, and so on.

### **BFS Algorithm:**

The BFS algorithm works by visiting a vertex v and marking it as visited. Then, for each unvisited neighbor u of v, we add u to a queue. Once all neighbors of v have been added to the queue, we take the next vertex from the front of the queue and repeat the process until all vertices have been visited. Here’s an outline of the steps involved in the BFS algorithm:

Start by putting any one of the graph’s vertices on top of a queue.
Take the front item of the queue and add it to the visited list.
Create a list of that vertex’s adjacent nodes.
Add the ones which aren’t in the visited list to the back of the queue.
Keep repeating steps 2 and 3 until the queue is empty.

### **BFS Pseudocode:**



```
BFS(G, start_vertex):
    let Q be a queue
    label start_vertex as discovered
    Q.enqueue(start_vertex)
    while Q is not empty do
        v = Q.dequeue()
        for all edges from v to w in G.adjacentEdges(v) do
            if w is not labeled as discovered then
                label w as discovered
                Q.enqueue(w)

```
In this pseudocode, G represents the graph and start_vertex represents the starting vertex for the search. The Q variable is a queue that keeps track of which vertices to visit next.

````python
# BFS implementation for Tree
# Diagram of tree is above.

tree = {'frank': ["man", "wurz", "kassel"], 'man':["karl"], 'wurz': ["num", "erf"], 'kassel':["mun"], 'karl': ["aus"], 'aus':[], 'num':["stu"], 'stu':[], 'erf':[], 'mun':[]}

def BFS_tree(tree, start_node):
    visited = [] # List to keep track of visited nodes
    queue = [start_node] # Initialize a queue with the start node

    while queue: # Loop until the queue is empty
        node = queue.pop(0) # Get the first node in the queue
        if node not in visited: # If the node has not been visited
            visited.append(node) # Add the node to the visited list
            key = tree[node] # Get the children of the current node
            for value in key: # Loop through the children
                queue.append(value) # Add each child to the queue
    return visited

print(BFS_tree(tree, 'wurz')) # Call the BFS function starting from the 'wurz' node
````

### **Applications of BFS:**
There are many applications of BFS, including finding the shortest path between two vertices, finding connected components in an undirected graph, and more. For example, finding connected components can be used to determine how many separate groups there are in a social network.

### **Complexity Analysis:**
The time complexity of BFS is O(V+E), where V is the number of vertices and E is the number of edges in the graph. This is because we need to visit every vertex and edge once during the search. The space complexity is O(V) due to the use of a queue to keep track of which vertices to visit next.
